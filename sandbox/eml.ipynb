
This is based on the [All elementary functions from a single operator](../docs/eml.pdf) paper.

```python
from math import exp, log

def eml(x,y):
    if y > 0:
        return exp(x) - log(y)
    return exp(x)

def e(x):
    return eml(x, 1)

def ln(x):
    return eml(1, eml(eml(1, x), 1))
```

```python
print(eml(1, 2))
print(eml(1,0))
```
